```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
```

```python
data = load_diabetes()
```

```python
X= data.data
```

```python
y = data.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Lr = LinearRegression()
```

```python
Lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score , root_mean_squared_error
```

```python
y_pred = Lr.predict(x_test)
```

```python
print("The r2 score is",r2_score(y_test, y_pred))
print("The root means square error" , root_mean_squared_error(y_test,y_pred))
```

```python
from sklearn.linear_model import Ridge
```

```python
R = Ridge(alpha=0.001)
```

```python
R.fit(x_train,y_train)
```

```python
y_pred1 = R.predict(x_test)
```

```python
print("The r2 score is",r2_score(y_test, y_pred1))
print("The root means square error" , root_mean_squared_error(y_test,y_pred1))
```

```python

m = 100
x1 = 5 * np.random.rand(m, 1) - 2
x2 = 0.7 * x1 ** 2 - 2 * x1 + 3 + np.random.randn(m, 1)

plt.scatter(x1, x2)
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
model = Pipeline([
        ('poly_feature', PolynomialFeatures(degree=16)) ,
        ('Ridge' , Ridge(alpha=0.001))
    ])
```

```python
def get_pred_ridge(x1,x2,alpha):
    model = Pipeline([
        ('poly_feature', PolynomialFeatures(degree=16)) ,
        ('Ridge' , Ridge(alpha=alpha))
    ])
    model.fit(x1,x2) 
    return model.predict(x1)
    
```

```python
alphas = [0, 20, 200]
cs = ['r', 'g', 'b']

```

```python
plt.figure(figsize=(10, 6))
plt.plot(x1, x2, 'b+', label='Datapoints')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x1, x2, 'b+', label='Datapoints')
for alpha, c in zip(alphas, cs):
    preds = get_pred_ridge(x1, x2, alpha)
    # Plot
    plt.plot(sorted(x1[:, 0]), preds[np.argsort(x1[:, 0])], c, label='Alpha: {}'.format(alpha))

plt.legend()
plt.show()
```
