```python
class custom_ridge_regression :
    def __init__(self,alpha = 1 , coef = 1 , intercept = 0) :
        
        self.alpha = alpha
        self.coef = coef
        self.intercept = intercept
        
    def fit(self, x_train , y_train):
       
        y_mean = np.mean(y_train)
        
        x_mean = np.mean(x_train)
        
        num = 0 
        deno = 0 
        for i in range(x_train.shape[0]) : 
            num = num + (y_train[i] - y_mean) * (x_train[i] - x_mean)
            deno = deno + ((x_train[i] - x_mean) ** 2)
            
        self.coef = num / (deno + self.alpha) 
        self.intercept = y_mean - self.coef * x_mean
        
    def predict(self , x_test) :
        return self.coef*x_test +  self.intercept
 
            
            
            
```

```python
cus_rig = custom_ridge_regression(alpha = 100)
```

```python
cus_rig.fit(X,y
            )
```

```python
print(cus_rig.coef)
cus_rig.intercept
```

```python
r2_score(y,cus_rig.predict(X))
```
