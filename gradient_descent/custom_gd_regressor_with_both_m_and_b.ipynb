```python
class GD_regressor_for_m_and_b:
    def __init__(self,coef = -120 , intercept = -120 ,epochs = 100 , learning_rate = 0.001):
        # initializing with any random value
        self.coef = coef
        self.intercept = intercept 
        self.epochs = epochs 
        self.learning_rate = learning_rate 
    
    def fit(self , x_train , y_train):
        
        for i in range(self.epochs):
            slope_for_b = (-2*np.sum(y_train - self.coef*x_train.ravel() - self.intercept))
            step_size_for_b = self.learning_rate*slope_for_b 
            self.intercept = self.intercept - step_size_for_b 
        
            slope_for_m = -2*np.sum(x_train.ravel()*(y_train - self.coef*x_train.ravel() - self.intercept)) 
            step_size_for_m = self.learning_rate*slope_for_m 
            self.coef = self.coef - step_size_for_m
            
            print("The m value is",self.coef)
            print("The intercept value is " , self.intercept)
            plt.scatter(x_train,y_train)
            plt.plot(x_train,self.coef * x_train + self.intercept , color = 'red')
            plt.show()
            
    def predict(self , x_test):
        return self.coef * x_test + self.intercept
        
         
```

```python
Lr_gd = GD_regressor_for_m_and_b()
```

```python
Lr_gd.fit(x_train,y_train)
```

```python
print(Lr_gd.intercept)
```

```python
print(Lr_gd.coef)
```

```python
y_pred = Lr_gd.predict(x_test=x_test)
```

```python
r2_score(y_test,y_pred)
```
