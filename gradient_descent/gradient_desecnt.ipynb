```python

from sklearn.datasets import make_regression
import numpy as np
     

X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color = 'red')
```

```python
lr.coef_
```

```python
lr.intercept_
```

# Lets apply our gradient decent

```python
b = 100
```

```python
learning_rate = 0.1
m = 78.35
```

# First iteration or first epochs

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color = 'red' , label = 'ols')
plt.plot(X,m*X + b , label = 'b = 100')
```

```python
# next iteration
```

```python
def get_slope(x,b):
    return -2 * np.sum(y - m*x.ravel() - b )
```

```python
new_b = b -  learning_rate*get_slope(X,b)
new_b
```

```python
m*X + b
```

```python
X[:,0]
```

```python
(m*X + new_b).shape
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color = 'red' , label = 'ols')
plt.plot(X,m*X + b ,color = '#00a65a' , label = 'b = 100')
plt.plot(X,m*X + new_b, color = '#A3E4D7' , label  = 'b_with_94')
plt.legend()
plt.show()
```

```python
# Third epoch
```

```python
new_bb = new_b - learning_rate*get_slope(X,new_b)
print(new_bb)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color = 'red' , label = 'ols')
plt.plot(X,m*X + b ,color = '#00a65a' , label = 'b = 100')
plt.plot(X,m*X + new_b, color = '#A3E4D7' , label  = 'b_with_40')
plt.plot(X,m*X + new_bb, color = '#A3E4D6' , label  = 'b_with_29.2')
plt.legend()
plt.show()
```

```python
new_bbb = new_bb - learning_rate*get_slope(X,new_bb)
print(new_bbb)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color = 'red' , label = 'ols')
plt.plot(X,m*X + b ,color = 'blue' , label = 'b = 100')
plt.plot(X,m*X + new_b, color = 'green' , label  = 'b_with_40')
plt.plot(X,m*X + new_bb, color = 'yellow' , label  = 'b_with_29.2')
plt.plot(X,m*X + new_bbb, color = 'grey' , label = 'b_with_26.75')
plt.legend()
plt.show()
```

```python
b = 100
epoch = 100
for i in epoch:
    b = b - learning_rate*get_slope(X,b)
    plt.plot(X,lr.predict(X),color = 'red' , label = 'OLS')
    plt.plot(X,m*X + b ,color ='#A3E4D7' , label = b)
    plt.legend()
    plt.show()
```
